```python
import os
import matplotlib.pyplot as plt

def readClass():
    f = open("/Users/anphong/Desktop/Full_Dataset/YOLOV4/train/_classes.txt", "r")
    content = f.readlines()
    f.close()
    res = []
    for line in content:
        line = line.split('\n')[0]
        res.append(line)
    return res

def getID(line):
    line = line.split(' ')[1]
    line = line.split(',')[4]
    return int(line)

classes = readClass()
cntClass = [0] * 9

f = open("/Users/anphong/Desktop/Full_Dataset/YOLOV4/train/train.txt", "r")
content = f.readlines()
f.close()
for line in content:
    id = getID(line)
    cntClass[id] += 1

print(classes)
print(cntClass)
print(sum(cntClass))
plt.bar(classes, cntClass)
```
