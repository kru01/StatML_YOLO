```python
import fiftyone
import pandas as pd
import os
import shutil
import warnings
warnings.filterwarnings("ignore")
```

```python
def clear_data():
    try:
        os.remove("/Users/anphong/fiftyone/open-images-v7/info.json")
    except: print("Already delete info.json")
    try:
        shutil.rmtree("/Users/anphong/fiftyone/open-images-v7/test/data")
    except: print("Already delete image folder")

    try:
        shutil.rmtree("/Users/anphong/Documents/School/ML/Crawling_Data/labels/test")
    except: print("Already delete label folder")
    # allFile = os.listdir("/Users/anphong/fiftyone/open-images-v7/train/data")
    # for file in allFile:
    #     os.remove("/Users/anphong/fiftyone/open-images-v7/train/data/" + file)
```

```python
def changeID(file, newID):
    f = open(file, "r")
    content = f.readlines()
    f.close()
    f = open(file, "w")
    for line in content:
        line = line.split(' ')
        newLine = str(newID)
        for i in range(1, 5): newLine += ' ' + line[i]
        f.write(newLine)
    f.close()
```

```python
def manipulate_id(newID):
    path = "/Users/anphong/Documents/School/ML/Crawling_Data/labels/test"
    allFile = os.listdir(path)
    for file in allFile:
        changeID(path + '/' + file, newID)
```

```python
def copy_file(animal):
    cnt = 0
    pathImg = "/Users/anphong/fiftyone/open-images-v7/test/data"
    pathLab = "/Users/anphong/Documents/School/ML/Crawling_Data/labels/test"
    allImages = os.listdir(pathImg)
    for image in allImages:
        label = image.split(".")[0] + ".txt"
        source = pathLab + "/" + label
        # An image may not have a label file
        if (not (os.path.isfile(source))):
            print(source + " is not existed")
            continue
        dest = "/Users/anphong/Documents/School/ML/Crawling_Data/New_dataset/test/labels/" +  animal + "_" + str(cnt) + ".txt"
        try: 
            shutil.move(source, dest)
        except:
            print(label + " already exists")

        source = pathImg + "/" + image
        dest = "/Users/anphong/Documents/School/ML/Crawling_Data/New_dataset/test/images/" + animal + "_" + str(cnt) + ".jpg"
        try: 
            shutil.move(source, dest)
        except:
            print(image + " already exists")
        
        
        cnt += 1
```

```python
def copy_labels():
    path = "/Users/anphong/Documents/School/ML/Crawling_Data/labels/test"
    allLabels = os.listdir(path)
    for label in allLabels:
        source = path + "/" + label
        dest = "/Users/anphong/Documents/School/ML/Crawling_Data/New_dataset/test/labels"
        try: 
            shutil.move(source, dest)
        except:
            print(label + " already exists")
```

```python
def download_animal(animal):
    classes = []
    classes.append(animal)
    dataset = fiftyone.zoo.load_zoo_dataset(
        "open-images-v7",
        split="test",
        label_types=["detections"],
        classes=classes,
        max_samples=400, #Test = 1
        dataset_name="Animal_test_" + animal
    )
    dataset.export(
        dataset_type=fiftyone.types.YOLOv5Dataset,
        classes=classes,
        split="test"
    )
```

```python
allClasses = ['Bear', 'Camel', 'Cat', 'Cattle', 'Chicken', 'Crab', 'Crocodile', 'Dog', 'Duck', 'Elephant', 'Fox', 'Frog', 'Hippopotamus', 'Kangaroo', 'Lion', 'Mouse', 'Pig', 'Rabbit', 'Sheep', 'Snail', 'Snake', 'Zebra']
allClasses.sort()
idx = 0
for animal in allClasses:
    print("Currently handle: " + animal)
    clear_data()
    download_animal(animal)
    manipulate_id(idx)
    copy_file(animal)
    idx += 1
```
